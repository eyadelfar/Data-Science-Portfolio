```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import math
from pandas_profiling import ProfileReport
```

```python
# Declaring some variables
titles_name,urls,pages_numbers,download_numbers = [] , [], [] , []
page = requests.get(f"https://www.pdfdrive.com/search?q=O%27reilly&pagecount=&pubyear=&searchin=&em=&page=1")
soup = BeautifulSoup(page.content,'html.parser')
url_numofresults = soup.find_all("div" , id = 'result-found')
url_pagesnum = math.ceil(int(url_numofresults[0].text.split(' ')[0].replace(',',''))/20)
pagessumreduce = url_pagesnum//2
url = "https://www.pdfdrive.com"
```

```python
# Start scraping

for i in range(1,pagessumreduce):

    page = requests.get(
        f"{url}/search?q=O%27reilly&pagecount=&pubyear=&searchin=&em=&page={i}"
    )
    soup = BeautifulSoup(page.content,'html.parser')

    titles , links = soup.find_all("h2") , soup.find_all("div" , class_="file-right")
    book_pages_num = soup.find_all("span" , class_="fi-pagecount")
    book_download_number = soup.find_all("span" , class_="fi-hit")


    for i in range(len(titles)):
        
        download = book_download_number[i].text.split()
        try:
            number = book_pages_num[i].text.split()
        except IndexError:
            number = ['Unknown']
            
        base = "https://www.pdfdrive.com" + links[i].find('a').get('href')

        
        titles_name.append(titles[i].text.strip())
        urls.append(base)
        pages_numbers.append(number[0])
        download_numbers.append(download[0])

print(len(titles_name))
print('-----------------------------')
print(len(urls))
print('-----------------------------')
print(len(pages_numbers))
print('-----------------------------')
print(len(download_numbers))
```

```python
df = pd.DataFrame({'title': titles_name,
                   'number_of_pages': pages_numbers,
                   'number_of_downloads': download_numbers,
                   'book_url': urls})
```
